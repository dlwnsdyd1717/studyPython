```python
# 딕셔너리 데이터 작성 (json 형식과 같다)
d = {"name" : "혼자 공부하는 데이터 분석"}
print(d['name'])
print(type(d))
```

```python
# 딕셔너리를 json 문자열로 변환하기 - json 라이브러리 (표준)
import json
# josn.dump() : 딕셔너리를 json 문자열로 변환
#  ensure_ascii=False : 유니코드를 변환하지 말자
d_str = json.dumps(d, ensure_ascii=False)
print(d_str)
print(type(d_str))
```

```python
# json 문자열을 딕셔너리로 변환
d2 = json.loads(d_str)
print(d2)
print(type(d2))
```

```python
# 여러개의 json 데이터를 딕셔너리 리스트로 입력
d4_str ="""
[
  {"name": "혼자 공부하는 데이터 분석", "author" : "박해선", "year" : 2022},
  {"name": "혼자 공부하는 머신러닝+딥러닝", "author" : "박해선", "year" : 2020}
]
"""
d4 = json.loads(d4_str)
print(d4[0]["name"])
print(d4)
```

```python
# 여러개의 json 문자열을 입력받은 딕셔너리 리스트를 pandas의 df에 저장해 보자.
from io import StringIO
import pandas as pd
pd.read_json(StringIO(d4_str))
# print(d4)
# print(type(d4)) # 데이터만 출력 , 줄 칸 형식 없어 맞지 않는다
```

```python
x_str = """
<book>
  <name>혼자 공부하는 데이터 분석</name>
  <author>박해선</author>
  <year>2022</year>
</book>
"""
```

```python
import xml.etree.ElementTree as et
book = et.fromstring(x_str)
```

```python
print(type(book))
```

```python
print(book.tag)
```

```python
# book.tag 안에 있는 tag들을 리스트로 만들어서 book_chileds에 입력
book_childs = list(book)
print(book_childs)
```

```python
name, author, year = book_childs
print(name.text)
print(author.text)
print(year.text)
# print(name.tag)
# print(author.tag)
# print(year.tag)
```

```python
x2_str ="""
<books>
  <book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
  </book>
  <book>
    <name>혼자 공부하는 머신러닝+딥러닝</name>
    <author>박해선</author>
    <year>2020</year>
  </book>
</books>
"""
```

```python
# books는 ElementTree.Element 타입이다
books = et.fromstring(x2_str)
print(books.tag)
```

```python
# books.findall('book') books 안에 있는 book이란 Element를 모두 찾는다
for book in books.findall('book'):
  name = book.findtext('name')    # book.findtext('name') book에서 name Element안에 있는 text를 찾아라
  author = book.findtext('author')
  year = book.findtext('year')
  print(name)
  print(author)
  print(year)
  print()
```
